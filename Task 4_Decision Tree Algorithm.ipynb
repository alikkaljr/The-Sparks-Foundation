# Task 4: To Explore Decision Tree Algorithm

__1) Importing necessary libraries__

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

__2) Importing Iris dataset__

```python
df=pd.read_csv("Iris.csv")
df.set_index(["Id"],inplace=True)
df.head()
```

__3) Visualizing the data__

```python
sns.pairplot(df,hue="Species")
```

__4) Splitting the data into training data and test data__

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]].values
y=df["Species"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

__5) Using Decision Tree Classifier__

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

__6) Making predictions__

```python
y_pred=dtree.predict(X_test)
```

__7) Checking how the model works__

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

__8) Plotting the Decision Tree__

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(13,13))
plot_tree(dtree,filled=True,feature_names=df.columns,class_names=df["Species"].unique().tolist())
```
